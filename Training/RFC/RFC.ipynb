```python
def fixFeatures(df):
    
    for column in df:
        df = df.rename(columns = {column : fixFeature(column)})
    
    return df
```

Carrega o banco de dados que contem somente fluxos benignos

```python
dfBenign = pd.read_feather('../../DataSets/CIC-DDoS2019/dfBenign')
```

```python
dfAttack = fixFeatures(dfAttack)
dfBenign = fixFeatures(dfBenign)
```

Descarta as features indesejadas

```python
dfAttack = dfAttack.drop('nothing', axis = 1)
dfBenign = dfBenign.drop('nothing', axis = 1)
dfAttack = dfAttack[features]
dfBenign = dfBenign[features]
```

Remove todos os fluxos benignos do banco de dados

```python
dfAttack = dfAttack[dfAttack['label'] == 'Syn']
```

Embaralha as aleatoriamente as linhas

```python
dfBenign = dfBenign.sample(frac=1)
```

Cria um banco de dados balanceado com o mesmo numero de ataques que legítimos

```python
df = pd.concat([dfAttack, dfBenign.sample(dfAttack.shape[0])])
```

Remoção dos inf e NaN

```python
# Troca todos 'inf' e '-inf' por NaN
df = df.replace([np.inf, -np.inf], np.NaN)
# Usa interpolação para remover os NaN
df = df.replace(np.NaN,0)
```

Divide o banco de dados

Sendo y os labels, X as features, train para o treinamento do modelo, e test para o test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('label',axis='columns'), df['label'], test_size=0.25)
```

#Configura os parâmetros do modelo que sera treinado

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
